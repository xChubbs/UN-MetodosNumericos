```python
# ----- Importamos las librerias ----- #

import matplotlib.pyplot as plt
import numpy as np
```

Para empezar, tomaré un ejemplo de referencia que se puede encontrar en la [documentación oficial](https://matplotlib.org "MATPLOTLIB"), En los comentarios voy a resaltar los elementos básicos para gráficar, y luego entremos un poco más en detalles de estilo!

```python

# ---------- Datos a gráficar ---------- #

# Definimos la nube de puntos y evaluamos
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# ----- Creamos los elementos del gráfico ----- #

# Generamos el elemento figura (fig) & axis (ax)
fig, ax = plt.subplots()

# Generamos el gráfico de la nube de puntos
ax.plot(t, s)

# Finalmente mostramos el gráfico
plt.show()
```

Demosle al gráfico anterior un poco más de formato, agreguemos elementos básicos que usamos en Matlab, como mallas, titulos, subtitulos, etc...

```python
# ---------- Datos a gráficar ---------- #

# Definimos la nube de puntos y evaluamos
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# ----- Creamos los elementos del gráfico ----- #

# Generamos el elemento figura (fig) & axis (ax)
fig, ax = plt.subplots()

# Generamos el gráfico de la nube de puntos
ax.plot(t, s)

# Agregamos titulos para cada axis y un titulo
ax.set(xlabel='Valor en X (x-val)',
       ylabel='Valor en y (y-val)',
       title='Mi primer gráfico')

# Agregamos una malla
ax.grid()

# Finalmente mostramos el gráfico
plt.show()
```

Ya si queremos ser más elegantes, podemos explorar formatos que hemos visto en los talleres en MATLAB, pero no hemos explorado a fondo, por ejemplo, definiremos una gráfica puntual desviada **(STEM)** y le agregaremos textos con formato y elementos para señalar otros detalles

```python
# ---------- Datos a gráficar ---------- #

# Definimos la nube de puntos y evaluamos
x = np.linspace(0.1, 4 * np.pi, 50)
t = np.linspace(0.1, 4 * np.pi, 100)

f = lambda x : np.exp(np.cos(x)) - 1

# -------- Creamos los elementos del gráfico -------- #

# Generamos el elemento figura (fig) & axis (ax)
fig, ax = plt.subplots()

# generamos linea suave de aproximación
ax.plot(t, f(t), '-.k', alpha= .6, label= "Aproximación")

# Agregamos nuestra linea de desviación
ax.plot([x[0], x[-1]], 2 * [0],'-k', linewidth= 1)

# Iterativamente generare lineas para unir los puntos
# Así damos efectos de desviación
for index in range(len(x)):

    # Gráficamos la linea desviada de cero
    ax.plot(2 * [x[index]], [0, f(x[index])],'-b', linewidth= 1)

# Gráficamos los puntos en el mismo color para generar contraste
ax.plot(x, f(x), '.b', label="Nube de puntos")

# Generemos una malla
ax.grid()

# Creemos una legenda
ax.legend(loc= 'best')

# Finalmente mostramos la imagen
plt.show()
```
